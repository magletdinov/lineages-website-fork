```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ROOT = Path().cwd()
DATA = ROOT / "data"
```

```python
other_results = {i.stem.split("_")[0]:pd.read_csv(i, sep="\t") for i in DATA.glob("*.tsv")}
```

```python
def groupby_and_aggregate(df):
    return df.groupby(by="lineage").aggregate({"lineage":"count"}).rename(columns={"lineage":"count"}).sort_values(by="count", ascending=False)
```

```python
def pie_plot(program_name):
    groupby_and_aggregate(other_results[program_name]).plot.pie(y='count', legend=False, autopct='%1.1f%%', figsize=(10, 10))
    plt.title(program_name, fontsize=20)
```

```python
groupby_and_aggregate(other_results['pangolin']).head()
```

```python
groupby_and_aggregate(other_results['vtrace'])
```

```python
pie_plot("pangolin")
```
